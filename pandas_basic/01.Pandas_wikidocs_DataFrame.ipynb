```python
# data라는 딕셔너리 생성
data = {
    'name':['Kim','Lee','Park'],
    'age':[24,27,34],
    'children':[2,1,3]
}
```

```python
# DataFrame에 data라는 딕셔너리 넣으면 자동 완성
df = pd.DataFrame(data)
```

```python
1. DataFrame 생성 df 변수에 할당
```

```python
import pandas as pd

food = {'food':['KFC','McDonald','SchoolFood'],
       'price':[1000,2000,2500],
       'rating':[4.5,3.9,4.2]}

df = pd.DataFrame(food)
df
```

```python
2. food 컬럼과 rating컬럼만 선택해 출력
```

```python
df[['food','rating']]
```

```python
3. food 컬럼명 place 컬럼명으로 변경********************
```

```python
df.rename(columns={'food':'place'})
df
```

```python
df.rename(columns={'food':'place'}, inplace=True)
df
```
